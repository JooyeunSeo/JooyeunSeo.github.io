# Pandas Schema & Code

```python
data = [[1, 8000, 'Jan'], [2, 9000, 'Jan'], [3, 10000, 'Feb'], [1, 7000, 'Feb'], [1, 6000, 'Mar']]
department = pd.DataFrame(data, columns=['id', 'revenue', 'month']).astype({'id':'Int64', 'revenue':'Int64', 'month':'object'})
```

```python
col_order = ["id", "Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
col_rename = {
    "id": "id",
    "Jan": "revenue_jan", "Feb": "revenue_feb", "Mar": "revenue_mar",
    "Apr": "revenue_apr", "May": "revenue_may", "Jun": "revenue_jun",
    "Jul": "revenue_jul", "Aug": "revenue_aug", "Sep": "revenue_sep",
    "Oct": "revenue_oct", "Nov": "revenue_nov", "Dec": "revenue_dec",
}

df = department.pivot(index='id', columns='month', values="revenue").reset_index()
df.columns.name = None
# x = df.reindex(columns=month_order).rename(columns=month_rename)
df
```
